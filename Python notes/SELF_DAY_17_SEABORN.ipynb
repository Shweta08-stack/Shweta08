SEABORN :-- Seaborn is a powerful libary for data visualization build on top of the matplotlib that provides high level and interactive statstical graphics.

```python
import numpy as np
```

```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
```

```python
import seaborn as sns
```

```python
titanic=sns.load_dataset("titanic")
```

```python
titanic.head(1)
```

# HISTOGRAM :-- for data distribution

x= continuous numerical

y= continuous numerical (frequency -count)

bins= converted into binary values(0,1)

```python
plt.figure(figsize=(6,5))
sns.histplot(data=titanic,x="age",bins=40,kde=True)  #bins;- cnverted graphs into parts like in 0-10 will give four parts,KDE= show the density
plt.title("age distribution")
```

# KDE PLOT :- IF you want only KDE plot.

```python

sns.kdeplot(data=titanic,x="age",color="purple",fill=True) #color will change only lin color and fill means fill the kde plot .
```

# SCATTER PLOT :- scatter plot is a type of data visualization that displays the points on two -dimensional graph.
-- Each points represnts the pair of value ,which one value is on the x-axis and another one is on y-axis.

x= continuous numerical

y= continuous numerical

## hue = to check male and female (we can check always with column name (eg:-"sex" , " survival", "pclass")

```python
plt.figure(figsize=(6,5))
sns.scatterplot(data=titanic,x="age",y="fare",hue="sex",palette='Set1')  # palette use for more than one color
plt.title("scatter plot of age vs fare")
```

## LINE PLOT :- We use for TRENDS (change the data second to second )(eg :-weather,treding etc)

based on conitinuous numerical value.

x= continuous numerical

y= coninuous numerical

```python
plt.figure(figsize=(5,5))
sns.lineplot(data=titanic,x="age",y="fare",hue="sex",palette="tab10")
plt.title("Line Plot")
```

## BAR PLOT :=  iT IS BASED ON THE CATEGORICAL AND NUMRICAL VALUE

x= categorical

y = numerical value


```python
plt.figure(figsize=(6,5))
sns.barplot(data=titanic,x="class",y="fare",hue='sex',palette="rocket")
plt.title("bar plot")

```

## COUNT PLOT :- A count plot is a type of plot used to display the number of counts of different categories in a dataset.

x= continous data count.



```python
titanic['sex'].value_counts()
```

```python
plt.figure(figsize=(6,5))
sns.countplot(data=titanic,x="sex",hue='sex',palette="hls")
plt.title("count plot of male and female")
```

##  BOX PLOT :-- TO FIND THE OUTLIER

```python
plt.figure(figsize=(6,5))
sns.boxplot(data=titanic,x="age",hue="survived",palette="Set1") HUE:- we can check both outlier at the same time.
plt.title("Find outlier")
```

## HEAT MAP :-- heatmaps are used when you need a clear, color-coded visualization to understand complex data .

```python
plt.figure(figsize=(6,5))
sns.heatmap(titanic.corr(numeric_only=True),cmap="coolwarm",annot=True)  ## annot :- gives the internal value in the box
plt.title("Heamap of the Titanic")







```

## VIOLIN PLOT :-- it is similar to box plot [find outlier] (only for design ,(we do not use--most time we use only the BOX PLOT)

```python
plt.figure(figsize=(6,5))
sns.violinplot(data=titanic,x="age",y="sex",palette="rocket")
plt.title("outlier of the data")
```
