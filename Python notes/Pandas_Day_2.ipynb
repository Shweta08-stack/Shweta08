```python
data = {
    'Employee': ['Alice', 'Bob', 'Charlie', 'David', 'Emma'],
    'Department': ['HR', 'IT', 'Finance', 'Marketing', 'Sales'],
    'Salary': [50000, 60000, 55000, 65000, 62000]
}
```

```python
data
```

```python
df = pd.DataFrame(data ,index = ['EMP01', 'EMP02', 'EMP03', 'EMP04', 'EMP05'])
```

```python
df['Bonus'] = df['Salary'] * 0.1
```

```python
new_df = df.where(df['Salary'] >= 60000)
```

```python
df['Salary_category'] = pd.cut(df['Salary'],
                               bins =[40000, 50000, 60000, 70000 ],
                               labels=['low', 'medium', 'high'])

#it use for feature engg. that mean to convert numerical feature to categorical
#bins = [0,18,32,60,100]
#labels = ['teen', 'Adult' , 'senior', 'OldAge']
```
