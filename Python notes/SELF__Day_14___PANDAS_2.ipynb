## GROUPBY ()[]:--
-- always use 1st paranthesis and 2nd square braces.

-- always use when you need more than one column calculation.

--- for more than 2 columns ,pass list in paranthesis braces

--- In sqaure braces output features or numerical in use.

```python
df.groupby("salary_category")['salary'].mean() ## first add sub column then add main column
```

```python
df.groupby('bonus')['salary'].mean()
```

```python
df.groupby('bonus')['salary'].mean()   ## for single value
```

```python
## UNSTACK :-- TO CHANGE THE FORMATTING
```

```python
df.groupby(["salary_category","emp"])['salary'].mean().unstack()
```
